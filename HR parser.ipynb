```python
import csv

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
file_name = 'Polar H7 B5DEB918_4AB5EEEC-5841-4E5D-9DEC-F7EC1310A568_1522173272206_hr-v1.csv'

reader = csv.reader(open(file_name, newline=''), dialect='excel')
```

```python
first_line = next(reader)
print(first_line)
```

```python
timestamps = []
HR_values = []
```

```python
for row in reader:
    print(row)
    
    if row[1]:
        timestamps.append(row[0])
        HR_values.append(row[1])
```

```python
print(timestamps)
```

```python
print(HR_values)
```

```python
plt.plot(timestamps[1000:1300], HR_values[1000:1300], 'ro')
plt.show()
```
